````python
strong_corr=corr_data[(corr_data>0.5)&(corr_data!=1)]
strong_corr
````

how many other features each variable is strongly correlated with (above 0.5)

````python

strong_corr = (corr_data > 0.5) & (corr_data < 1)
strong_corr.sum()
````

````python
corr_data
````

#EXTRACTING CORRELATED PAIRS

masking upper triangles (excluding duplicates and self correlations)i-e excluding diagonal

````python
str_corr=corr_data.stack()
str_corr
````

````python
mask=np.triu(np.ones_like(corr_data,dtype=bool),k=1)
mask
````

corr.where(mask): Replaces everything not in the upper triangle with NaN.

applying mask to hide upper traingle

````python

masked_corr=pd.DataFrame(corr_data.where(mask),index=corr_data.index,columns=corr_data.columns)
masked_corr
````

stack mask correlation matrix i-e turning into series of pairs

````python

strong_corr=masked_corr.stack()
strong_corr
````

EXTRACTING STRONG RELATIONSHIPS

setting a threshold for stronng relationship

````python
threshold=0.5
````

filter for very strong correlations(absvalue>threshold)

````python
very_strong_corr = strong_corr[strong_corr.abs() > threshold]
very_strong_corr
````

Print correlated pairs

````python
for pair in strong_corr.index:
  col1,col2=pair
  value=strong_corr[pair]
  print(f"{col1} <-> {col2} ={value:.2f}")
````

##GROUP BY


This approach helps to analyze how price varies with different ranges of area, providing valuable insights into the relationship between these two variables.
```
`0-4000 sq ft:` The average price is  approximately $3.60 million.

`4000-6000 sq ft:` The average price increases to about $4.95 million.

`6000-8000 sq ft:` The average price further rises to around $6.06 million.

`8000+ sq ft:` The average price peaks at approximately $6.17 million.

This indicates a trend where larger areas tend to have higher average prices

````python
#creating price range column
# Create price bins
bins = [0, 3000000, 5000000, 8000000, df['price'].max()]
labels = ['Low', 'Medium', 'High', 'Luxury']
df['price_range'] = pd.cut(df['price'], bins=bins, labels=labels)
````

````python
df.head(2)
````

````python
# creating a new column area_category
bins=[0, 4000, 6000, 8000, 20000]
labels = ['0-4000', '4000-6000', '6000-8000', '8000+']
df["area_category"]=pd.cut(df["area"],bins=bins,labels=labels)
````

#`Storytelling `

##Summary-----Housing Market


`Insights`

In our housing dataset, the market reveals a story of size, features, and value.

* Most homes sit comfortably in the middle market â€” midsized plots with mid-range prices â€” forming the heart of the market activity. The 0-4000 sq ft segment dominates in number, while the 8000+ sq ft category is rare, almost exclusive territory.

* As we move up in area size, prices climb steadily, and luxury features start appearing. Larger plots are more likely to boast basements, air conditioning, and preferred locations, making them clear markers of high-end living. Interestingly, the low-price range fades away almost entirely beyond 4000 sq ft, replaced by mid-range and luxury tiers.

* Bathrooms emerge as a silent but strong influencer. Homes with more bathrooms command noticeably higher prices â€” both on average and at the median â€” suggesting buyers pay a premium for comfort and convenience. Properties with four bathrooms often sit in the top pricing tier, some matching their maximum price to their median, signaling consistent high value.

* The 4000 to 6000 sq ft category takes the crown for total sales value, showing its sweet spot between exclusivity and demand.

* Yet, the market is not without its quirks. Outliers â€” massive plots with staggering prices â€” pull the mean upwards, making median prices a truer reflection of what a â€œtypicalâ€ buyer might expect to pay. Bedrooms correlate with higher prices, but not perfectly; other features like stories, main road access, and furnishing status often matter just as much.

* For buyers, this means: compare median prices for similar-sized, similarly equipped homes to get a fair view of value.

   With the wide spread of sizes, amenities, and prices in the market, anyone can find a property that fits their budget range â€” from practical mid-market homes to exclusive luxury estates.

* For sellers,price your property in line with others of the same size and features to stay competitive.
